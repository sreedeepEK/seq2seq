```python
'''  
we need to convert from Unicode to ASCII. This is to limit the RNN input layers
'''

import string
import unicodedata

allowed_characters = string.ascii_letters + " .,;\'_"
n_letters = len(allowed_characters)

# https://stackoverflow.com/a/518232/2809427 

''' 
Basically, Take a string that might contain accented or non-ASCII characters 
(like ‘é’, ‘ü’, ‘ñ’, ‘ç’, or even non-Latin scripts like ‘Ж’ or ‘你’) 
and simplify it to plain English alphabet characters (A–Z, a–z) and a few 
punctuation marks.
'''

def unicodeToAscii(s):
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'
        and c in allowed_characters
    )
```

```python
print (f"converting 'Ślusàrski' to {unicodeToAscii('Ślusàrski')}")
```
